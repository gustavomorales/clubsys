**Question #1:** How many unique orders are in the orders.csv file?

**Question #2:** What is the average number of items per order (rounded to two decimal places)?

**Question #3:** What is the highest number of items per order?

**Question #4:** What is the number of orders placed in October 2021?

**Question #5:** Which customer spent the most amount of money in 2021?

**Question #6:** Historically, what is the best month for sales?

```python
# open file and init vars
from os.path import exists
import csv
import datetime
import calendar

# check if file can be found and open the file
datafile = "./sample_data/orders.csv"
if not exists(datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

items_per_order={}
index = 0
orders_in_october = 0
amount_by_customer_in_2021 = {}
sales_by_month = {}
with open(datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  for order_line in csvreader:
    index += 1
    if index == 1 : continue
    order_id = order_line[1]
    customer_id = order_line[0]
    order_total = float(order_line[3])
    order_date_value = order_line[2]
    if order_date_value != "":
      order_date = datetime.datetime.strptime(order_date_value, '%Y-%m-%d %H:%M:%S.%f')
      order_month = order_date.month
      order_year = order_date.year
      if order_month not in sales_by_month: sales_by_month[order_month] = order_total
      else: sales_by_month[order_month] += order_total
      if order_month == 10: orders_in_october += 1
      if order_id not in items_per_order:
        items_per_order[order_id] = 1
        if order_year != 2021: continue
        if customer_id not in amount_by_customer_in_2021: amount_by_customer_in_2021[customer_id] = order_total
        else: amount_by_customer_in_2021[customer_id] += order_total
      else: items_per_order[order_id] += 1


total_items = 0
hightest_items = 0
for order_id in items_per_order:
  total_items += items_per_order[order_id]
  if items_per_order[order_id] > hightest_items: hightest_items=items_per_order[order_id]

unique_order_count = len(items_per_order.keys())
average_items_per_order = round(total_items / unique_order_count, 2)


customer_spend_most_id = ""
higher_amount = 0

for amount_by_customer in amount_by_customer_in_2021:
  if amount_by_customer_in_2021[amount_by_customer] > higher_amount:
    customer_spend_most_id = amount_by_customer
    higher_amount = amount_by_customer_in_2021[amount_by_customer]

# customer file
customer_datafile = "./sample_data/customers.csv"
if not exists(customer_datafile):
  raise SystemExit("You should run the first code cell and download the dataset files!")

customer_spend_most_full_name = ""
with open(customer_datafile, 'r') as fl:
  csvreader = csv.reader(fl, delimiter=',')
  for customer_line in csvreader:
    index += 1
    if index == 1 : continue
    customer_id = customer_line[0]
    if customer_id == customer_spend_most_id:
      firstname =customer_line[1]
      lastname = customer_line[2]
      customer_spend_most_full_name = firstname + " " + lastname
      break
    else: continue


sales_by_month_sorted_desc = dict(sorted(sales_by_month.items(), key=lambda item: item[1], reverse=True))
best_month_sales = list(sales_by_month_sorted_desc)[0]

# evaluate the contents of the file
print("The unique orders are %s" % unique_order_count)
print("The average items per orders is %s" % average_items_per_order)
print ("The hightest numbers of items in a order is %s" %hightest_items)
print ("The number of order placed in October are %s" %orders_in_october)
print ("The customer spend the most amount in 2021 was %s" % customer_spend_most_full_name)
print ("The best month for sales is %s" %calendar.month_name[best_month_sales])
```
